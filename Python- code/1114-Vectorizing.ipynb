## Word2Vec- vectorizing- 

```python
import pandas as pd 


PATH = './data_out/'
CSV = 'train_clean.csv'

train_data = pd.read_csv(PATH+CSV)

clauses = list(train_data['clauses'])
label = list(train_data['label'])

sentences =[]
for clause in clauses:
    sentences.append(clause.split())
```

```python
sentences
```

```python
# word2Vec - vectorizing hyperparameter 

num_features = 300
min_word_count=10
num_workers=4
context=10
downsampling=1e-3

```

```python
import logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level= logging.INFO)
```

```python
from gensim.models import word2vec
print("training model...")
model = word2vec.Word2Vec(sentences, 
                          workers=num_workers, 
                          size= num_features, 
                          min_count=min_word_count, 
                          window= context, 
                          sample = downsampling)
```

```python
model_name = '1114-300f-10minwords-10context'
model.save(model_name)
```
